#S&P 500 July Stock Data

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

start_date = '2024-07-16'
end_date = '2024-07-31'

stock_name = "^GSPC"
ticker = yf.Ticker(stock_name)
stock_data = ticker.history(start=start_date, end=end_date)

stock_data

print(stock_data)

# Plot the Adjusted Closing Price
plt.figure(figsize=(12, 6))
stock_data['Close'].plot()
plt.title(f"{stock_name} Adjusted Closing Price")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid()
plt.show()

# Plot the Volume
plt.figure(figsize=(12, 6))
stock_data['Volume'].plot()
plt.title(f"{stock_name} Volume")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid()
plt.show()
```

#Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Coca Cola', 'McDonalds', 'Comcast', 'FOX', 'American Airlines','Delta Airlines', 'Topgolf', 'Footlocker']
values = [coke_pdiff, mcd_pdiff, comcast_pdiff, fox_pdiff, aal_pdiff, dal_pdiff, topgolf_pdiff, footlocker_pdiff]


x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects = ax.bar(x - width/2, values, width, label='Values')


ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Chart Example')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
